# **Pandas**

It has functions for analyzing, cleaning, exploring and manipulating data.

Pandas are also able to delete rows that are not relevant, or contains wrong values.


**USE of Pandas:**

* Easily handles missing data.
* It provides an efficient way to slice the data
* It provides a flexible way to merge, concat or reshape the dta.
* It includes a powerful time series tool to work with.
* It uses Series for one-Dimensional data structure and DataFrame for the multi-dimensional data structure.



**Importing Pandas**

```python
import pandas as pd
```

# **Advantages of Pandas**
* Fast and efficient for manipulating and analyzing data. 
* Data from different file objects can be loaded. 
* Easy handling of missing data (represented as NaN) in floating point as well as non-floating point data 
* Size mutability: columns can be inserted and deleted from DataFrame and higher dimensional objects 
* Data set merging and joining. 
* Flexible reshaping and pivoting of data sets 
* Provides time-series functionality. 
* Powerful group by functionality for performing split-apply-combine operations on data sets.


## **Pandas Dataframes**

Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes known as rows and columns. a dataframe is a collection of series that can be used to analyse the data.  DataFrame can be created from the lists, dictionary, and from a list of dictionaries etc.

```python
import numpy as np

data=np.array(['a','b','c','d'])
s=pd.Series(data)  #Pandas series
print(s)
```

```python
print(data)
print(type(data))
print(type(s))
```

```python
data=np.array(['a','b','c','d'])
s=pd.Series(data,index=[100,101,102,103])
print(s)
```

```python
s=pd.Series(data,index=[100,"Abhijit",102,103])
print(s)
```

```python
data = {'a':0, 'b':1, 'c':2, 'd':3}
s=pd.Series(data)
print(s)
```

```python
print(s.unique())
print(set(s))
```
